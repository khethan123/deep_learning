            </li>
            <ul>
                <li>
                    To change the font-family, first you should surround your text in a html tag such as <code>span</code>, <code>p</code> and <code>div</code> and then you have to add the css <code>style</code> property inside the html tag. After this you can change the family of font by <code>font-family</code> property. Lets look at an example to make sense of all this:<br><br>
                    &lt;span style="font-family:cursive;"&gt; Love for all, Hatred for none &lt;/span&gt;
                    <br><br>
                    <span style="font-family:cursive;">Love for all, Hatred for none</span>
                </li>
            </ul>
            <br>
            <li>
                <b>Change Font Color</b>
            </li>
            <ul>
                <li>
                    To change the font-color, first you should surround your text in a html tag such as <code>span</code>, <code>p</code> and <code>div</code> and then you have to add the css <code>style</code> property inside the html tag. After this you can change the color of font by <code>color</code> property. Lets look at an example to make sense of all this:<br><br>

            </ul>
